## Top 10 countries by population in 2023

```python
top_10_countries = df.sort_values(by = "2023 population", ascending = False).head(15)  
```

```python
top_10_countries
```

```python
sns.barplot(x = "2023 population", y = "country", data = top_10_countries) 
plt.title("Top 10 Countries by Population in 2023") 
plt.xlabel("Population") 
plt.ylabel("Country") 
plt.show() 
```

```python
df.rename(columns = {"density (km²)": "density (km2)"}, inplace = True) 
```

```python
df.columns
```

```python
sns.scatterplot(x= "2023 population", y= "density (km2)", data= df)
plt.title("Population and Destiny Scatter plot")
plt.xlabel("2023 Population") 
plt.ylabel("Density") 
plt.show() 
```

```python
df.columns
```

```python
df.rename(columns = {"area (km²)": "area (km2)"}, inplace = True) 
```

```python
df.columns
```

```python
df.groupby("continent")["area (km2)"].sum().sort_values(ascending = False) 
```

```python
sns.violinplot(x = "continent", y= "density (km2)", data= df, hue = "continent", palette = "viridis")
plt.title("Population Density Distribution Across Continents")
plt.xlabel("Continent") 
plt.ylabel("Population Density") 
plt.show() 
```

```python
sns.boxplot(x = "continent", y = "density (km2)", data = df, hue = "continent", palette = "muted", showfliers = False)  
plt.title("Population Density Distribution Across Continents (Without Outliers)") 
plt.xlabel("Continent") 
plt.ylabel("Population Density") 
plt.show() 
```
